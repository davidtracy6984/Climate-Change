```python
#set up paths
path = 'csvfiles\\'
```

```python
#dummy_df = pd.read_csv('StormEvents_details-ftp_v1.0_d1950_c20170120.csv')
dummy_df = pd.read_csv(path + 'StormEvents_details-ftp_v1.0_d1951_c20160223.csv')
dummy_df.head()
```

```python
all_files = glob.glob(path + "/*.csv")
```

```python
#Now I want to make three lists, one for each type of data
details_list = []
fatalities_list = []
locations_list = []
for filename in all_files:
    if "details" in filename:
        df = pd.read_csv(filename, index_col=None, header=0)
        details_list.append(df)
    elif "fatalities" in filename:
        df = pd.read_csv(filename, index_col=None, header=0)
        fatalities_list.append(df)
    elif "locations" in filename:
        df = pd.read_csv(filename, index_col=None, header=0)
        locations_list.append(df)
```

```python
len(locations_list)
```

```python
details_df = pd.concat(details_list, axis=0, ignore_index=True)
```
