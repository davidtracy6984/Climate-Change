```python
#Import the libraries that will be used
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#set up the pathway for the water level file
path = 'Data\\'

#Making a datframe out of the file
ice_raw = pd.read_csv(path+'ice_coverage.csv')
ice_raw.head()
```

```python
#For Some reason the Superior in the rawe file has a space infront of it, best just to get ride of it
ice_raw.columns = ['year', 'Superior', 'Michigan', 'Huron', 'Erie', 'Ontario',
       'All_Lakes']
```

```python
#Next I want to setup my variables
x = ice_raw['year']
y = ice_raw['Superior']
circle_radius = 20

#Then the start of the plot
fig_ice_s, (ax1) = plt.subplots(1,1,figsize=(15,5))

ax1.plot(x,y,c='b',label='Ice Coverage')
ax1.plot(1977, 96, 'o',ms=circle_radius,color='r',mfc='none',mew=2)
ax1.plot(1982, 85, 'o',ms=circle_radius,color='r',mfc='none',mew=2)
ax1.plot(1984.95, 83, 'o',ms=circle_radius,color='r',mfc='none',mew=2)
ax1.plot(1989, 82, 'o',ms=circle_radius,color='r',mfc='none',mew=2)
ax1.plot(2014.5, 95, 'o',ms=circle_radius*1.5,color='r',mfc='none',mew=2)
ax1.plot(2019, 96, 'o',ms=circle_radius,color='r',mfc='none',mew=2)
ax1.set(xlabel='Year',ylabel='Ice coverage',title='Lake Superior')
ax1.grid(linestyle = '-.')
```

```python
#Data for ice coverage comes from https://www.glerl.noaa.gov/data/ice/#historical
#Years that polar vortex storm occured from https://www.weather.gov/safety/cold-polar-vortex
```
